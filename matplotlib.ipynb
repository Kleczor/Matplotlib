```python
import numpy as np
import matplotlib.pyplot as plt
import math 

x = np.arange(0.0, 2.0, 0.01)
y1 = []
y2 = []
for num in x:
    y1.append(np.cos(2.0*np.pi*num))
    y2.append(np.cos(2.0*np.pi*2.0*num)*math.exp(num))
plt.plot(x,y1,'r:',x,y2,'g')
plt.legend(('tłumione','harmonijne'))
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.title('Subploty ')
plt.grid(True)
plt.show()
```
